```python
# Re-importing pandas as the code execution state was reset
import pandas as pd

# Load the CSV file again to reprocess the data
df = pd.read_csv('jailbreaks_by_goal.csv')

# Define the mediums
mediums = {
    "Poem": ["poem"],
    "Song": ["song"],
    "Tutorial Video": ["tutorial video"],
    "Message": ["message"],
    "Blog Post": ["blog post"],
    "App": ["app"],
    "Website": ["website"],
    "Social Media": ["social media campaign", "social media post"],
    "Description": ["describe in detail"],
    "Instructions": ["instructions", "instruct", "guide", "guidance"]
}

# Function to determine the medium based on the goal description
def determine_medium(goal):
    for medium_name, keywords in mediums.items():
        if any(keyword in goal.lower() for keyword in keywords):
            return medium_name
    print(goal)
    return "Other"

# Assign mediums to each row
df['Medium'] = df['goal'].apply(determine_medium)

# Group by medium and count True/False occurrences
medium_counts = df.drop(columns='goal').groupby('Medium').agg(lambda x: x.value_counts().to_dict())

medium_counts.to_csv('medium_counts_2.csv')
```
